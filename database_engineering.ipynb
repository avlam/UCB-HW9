```python
# Dependencies
import pandas as pd
from datetime import datetime
import os
# SQL Alchemy
from sqlalchemy import Column, Float, Integer, String, Date
from sqlalchemy import create_engine
from sqlalchemy.orm import Session 
from sqlalchemy.ext.declarative import declarative_base
Base = declarative_base()
```

```python
# Create class for Measurements
class Measurement(Base):
    __tablename__ = 'measurement'
    measure_id = Column(Integer, primary_key = True)
    station = Column(String)
    date = Column(Date)
    prcp = Column(Float)
    tobs = Column(Float)
```

```python
# Create class for Stations
class Station(Base):
    __tablename__ = 'station'
    station_id = Column(Integer, primary_key = True)
    station = Column(String)
    name = Column(String)
    latitude = Column(Float)
    longitude = Column(Float)
    elevation = Column(Float)
```

```python
# Create Engine and Pass in MySQL Connection
dialect = 'sqlite'
database = 'hawaii'
engine = create_engine(f'{dialect}:///{database}.{dialect}')
Base.metadata.create_all(engine)
session = Session(bind = engine)
```

```python
# clean data files
clean_dir = 'clean_data'
station_file = 'clean_hawaii_stations.csv'
measurement_file = 'clean_hawaii_measurements.csv'
```

```python
# function to convert date string imported from csv to datetime object
def convert_str_to_date(date_str):
    out = datetime.strptime(date_str,'%Y-%m-%d')
    return out
```

```python
# import and add station data
station_data = pd.read_csv(os.path.join(clean_dir,station_file),index_col = 0)
# station_data.head()
enter_to_database = []
for entry in station_data.index:
    enter_to_database.append(Station(**station_data.loc[entry]))

# import and add measurement data
measure_data = pd.read_csv(os.path.join(clean_dir,measurement_file),index_col = 0)
measure_data['date'] = measure_data['date'].apply(convert_str_to_date)
# station_data.head()
for entry in measure_data.index:
    enter_to_database.append(Measurement(**measure_data.loc[entry]))
```

```python
# Commit entries to database
session.add_all(enter_to_database)
session.commit()
```
